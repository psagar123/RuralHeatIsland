```python
# First process geo_em* data
# Provide the path to your file
file_path_0 = '/content/drive/MyDrive/Google_Drive_UTmail/SDSU/python_colab_data/geo_em.d03_10m.nc'  # Update with the actual path

# Open the NetCDF file
data_0 = nc.Dataset(file_path_0, mode='r')
print(data_0.dimensions.keys()) # dimension names
print(data_0.variables.keys()) # variable names
# Access the variable
lai_0 = data_0.variables['LAI12M'][:]
lat_0 = data_0.variables['XLAT_M'][:]
lon_0 = data_0.variables['XLONG_M'][:]
print(lai_0.shape)


lai_0_av_lon = np.nanmean(lai_0, axis=3)
print(lai_0_av_lon.shape)
lai_0_av_lon_lat = np.nanmean(lai_0_av_lon, axis=2)
print(lai_0_av_lon_lat.shape)

lai_0_av_lon_lat_flatten = np.ravel(lai_0_av_lon_lat)


file_path_1 = '/content/drive/MyDrive/Google_Drive_UTmail/SDSU/python_colab_data/geo_em.d03_30s.nc'  # Update with the actual path

# Open the NetCDF file
data_1 = nc.Dataset(file_path_1, mode='r')
print(data_1.dimensions.keys()) # dimension names
print(data_1.variables.keys()) # variable names
# Access the variable
lai_1 = data_1.variables['LAI12M'][:]

lai_1_av_lon = np.nanmean(lai_1, axis=3)
print(lai_0_av_lon.shape)
lai_1_av_lon_lat = np.nanmean(lai_1_av_lon, axis=2)
print(lai_1_av_lon_lat.shape)
lai_1_av_lon_lat_flatten = np.ravel(lai_1_av_lon_lat)


file_path_2 = '/content/drive/MyDrive/Google_Drive_UTmail/SDSU/python_colab_data/geo_em.d03_500m.nc'  # Update with the actual path

# Open the NetCDF file
data_2 = nc.Dataset(file_path_2, mode='r')
print(data_2.dimensions.keys()) # dimension names
print(data_2.variables.keys()) # variable names
# Access the variable
lai_2 = data_2.variables['LAI12M'][:]

lai_2_av_lon = np.nanmean(lai_2, axis=3)
print(lai_2_av_lon.shape)
lai_2_av_lon_lat = np.nanmean(lai_2_av_lon, axis=2)
print(lai_0_av_lon_lat.shape)
lai_2_av_lon_lat_flatten = np.ravel(lai_2_av_lon_lat)
```

```python
from datetime import datetime, timedelta
import calendar

t1 = datetime(2020, 1, 1)
t2 = datetime(2020, 12, 1)

# Generate a list of first days of each month
t = [datetime(year, month, 1) for year in range(t1.year, t2.year + 1) for month in range(1, 13)]
print(t)
tt = np.array(t)
print(tt)
```

```python
# Now lets plot the timeseries from the sptially-averaged data

import matplotlib.pyplot as plt
import numpy as np

fig, (ax1) = plt.subplots(1, 1, figsize=(16, 4))
plt.subplots_adjust(hspace=0.5)

ax1.plot(tt, lai_0_av_lon_lat_flatten, color=[1, 0, 1], linewidth=2.5, label='Coarse 10 arc-min')
ax1.plot(tt, lai_1_av_lon_lat_flatten, color=[0, 0, 1], linewidth=2.5, label='Fine 30 arc-sec')
ax1.plot(tt, lai_2_av_lon_lat_flatten, color=[0, 1, 0], linewidth=2.5, label='New 0.5 km')

ax1.set_title('Average LAI over the study domain (d03)')
ax1.set_xlabel('Date')
ax1.set_ylabel('LAI (m$^2$/m$^2$)')
ax1.tick_params(axis='both', labelsize=12)
ax1.grid(True)
ax1.set_yticks(np.arange(0.1, 0.5 + 0.1, 0.1))
ax1.legend(loc='upper right')
ax1.set_xlim([t1, t2])

plt.savefig('timeseries_lai_comparision.png', bbox_inches='tight', dpi=300)
```
